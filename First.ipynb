```python
# Plotting the vector 
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1, color='r'
        )
plt.quiver(0, 0, 4, -5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
# Addition of vectors
vector1 = np.array([4, 5])
vector2 = np.array([4, -5])
vector = vector1 + vector2
vector
plt
```

```python
a = np.array([6,7,4])
b = np.array([2,3, 1])
a_dot_b= np.dot(a,b)
```

```python
a = np.array([6,7,4])
b = np.array([2,3, 1])
a_dot_b= np.cross(a,b)
```

```python
a = np.array([2,5])
b = np.array([8,6])
mag =np.sqrt(sum(a**2))
proj_of_a_on_b = (np.dot(a,b)/mag**2)*a
```
